```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
import statsmodels.api as sm
from statsmodels.tsa.stattools import coint
from src.pairs_trading_utils import (
    get_data, test_cointegration, calculate_spread,
    compute_rolling_zscore, generate_signals,
    backtest_strategy, evaluate_performance
)


plt.style.use("seaborn-v0_8-darkgrid")
```

```python
data = yf.download(["KO", "PEP"], start="2019-01-01", end="2024-01-01")
data = data[["Close"]]
data.columns = ["KO", "PEP"]
data.head()
```

```python
#Plot KO vs PEP
data.plot(figsize=(14, 6), title='KO vs PEP Adjusted Close Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()

#Compute and plot spread
spread = data['KO'] - data['PEP']

spread.plot(figsize=(14, 6), title="Price Spread (KO - PEP)")
plt.xlabel('Date')
plt.ylabel('Price Difference')
plt.axhline(spread.mean(), color='black', linestyle='--', label='Mean')
plt.legend()
plt.show()
```

```python
#Z-score function
def zscore(series):
    return (series - series.mean()) / series.std()

#Compute and plot z-score of spread
zscore_spread = zscore(spread)

zscore_spread.plot(figsize=(14, 6), title="Z-Score of Price Spread (KO - PEP)")
plt.axhline(0, color='black', linestyle='--', label="Mean")
plt.axhline(1, color='red', linestyle='--', label="+1 Std")
plt.axhline(-1, color='green', linestyle='--', label="-1 Std")
plt.legend()
plt.xlabel('Date')
plt.ylabel('Z-score')
plt.show()
```

```python
#Initializing empty signal series aligned with z-score index
rolling_signals = pd.Series(index=rolling_zscore.index, dtype='float64')

#Generate trading signals based on z-score logic
rolling_signals[rolling_zscore > 1] = -1  #Short spread (KO overvalued)
rolling_signals[rolling_zscore < -1] = 1  #Long spread (KO undervalued)
rolling_signals[(rolling_zscore >= -1) & (rolling_zscore <= 1)] = 0  #Neutral zone
```

```python
plt.figure(figsize=(14, 6))
rolling_signals.plot(label='Rolling Signal', lw=1)
plt.title('Adaptive Trading Signal Based on Rolling Z-Score Thresholds')
plt.axhline(0, color='black', linestyle='--', label='Neutral')
plt.axhline(1, color='green', linestyle='--', label='Long Entry')
plt.axhline(-1, color='red', linestyle='--', label='Short Entry')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Signal')
plt.show()
```

```python
#Compute daily returns
returns = data.pct_change()

#Develop strategy returns
strategy_returns = rolling_signals.shift(1) * (returns['KO'] - returns['PEP'])

#Cumulative returns over time
cumulative_returns = (1 + strategy_returns).cumprod()

#Plot cumulative strategy return
plt.figure(figsize=(14, 6))
cumulative_returns.plot(label='Strategy Return', color='blue')
plt.title('Cumulative Returns of KOPEP Pairs Trading Strategy')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Engle-Granger cointegration test
score, pvalue, _ = coint(data['KO'], data['PEP'])
print(f'Cointegration Test p-value: {pvalue:.4f}')
```

```python
window = 60
rolling_mean = spread.rolling(window=window).mean()
rolling_std = spread.rolling(window=window).std()

#Adaptive z-score
rolling_zscore = (spread - rolling_mean) / rolling_std

plt.figure(figsize=(14, 6))
rolling_zscore.plot(title=f"Rolling Z-Score of Spread (Window={window})")
plt.axhline(0, color='black', linestyle='--', label='Mean')
plt.axhline(1, color='red', linestyle='--', label='+1 Std')
plt.axhline(-1, color='green', linestyle='--', label='-1 Std')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Rolling Z-Score')
plt.show()
```

```python
# Risk-free rate (daily, from 4.5% annual)
rf_annual = 0.045
rf_daily = rf_annual / 252

excess_returns = strategy_returns - rf_daily
total_return = cumulative_returns.iloc[-1] - 1

#Sharpe Ratio
sharpe_ratio = np.sqrt(252) * excess_returns.mean() / excess_returns.std()

#Max Drawdown
rolling_max = cumulative_returns.cummax()
drawdown = cumulative_returns / rolling_max - 1
max_drawdown = drawdown.min()

# CAGR (Compounded Annual Growth Rate)
n_days = len(strategy_returns.dropna())
cagr = (1 + total_return) ** (252 / n_days) - 1

# Output
print(f"Total Return       : {total_return:.2%}")
print(f"Sharpe Ratio       : {sharpe_ratio:.2f}")
print(f"Max Drawdown       : {max_drawdown:.2%}")
print(f"Annualized Return  : {cagr:.2%}")
```

```python
#Parameters
tickers = ["KO", "PEP"]
start_date = "2019-01-01"
end_date = "2024-01-01"
zscore_window = 60

#Pipeline
data = get_data(tickers, start=start_date, end=end_date)
spread = calculate_spread(data, "KO", "PEP")
zscore = compute_rolling_zscore(spread, window=zscore_window)
signals = generate_signals(zscore)

#Backtest
returns = data.pct_change()
strategy_returns = backtest_strategy(signals, returns[["KO", "PEP"]])
cumulative_returns = (1 + strategy_returns).cumprod()

#Performance
results = evaluate_performance(strategy_returns, cumulative_returns)

#Display Results
for key, val in results.items():
    print(f"{key}: {val}")
```
